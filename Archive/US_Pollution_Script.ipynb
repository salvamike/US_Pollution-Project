```python
# Import Libraries
import pandas as pd
import glob
import os
```

```python
# Create Path
path = "Data_Sources"
```

```python
# Test Directory of Files in Path
dir_list = os.listdir(path)
print(dir_list)
```

```python
# Test Data Source
twelve = pd.read_csv('Data_Sources/annual_aqi_by_cbsa_2012.csv')
```

```python
# Get all files from path
all_files = glob.glob(path + "/*.csv")
print(all_files)
```

```python
# Read each file into DataFrame
df_list = (pd.read_csv(file) for file in all_files)
```

```python
# Concatenate All Files into one Dataframe
full_df = pd.concat(df_list, axis=0)

# Check Head for 2012 Data
full_df.head(10)
```

```python
# Check Tail for 2022 Data
full_df.tail(10)
```

```python
# Check all years of data (2012-2022)
full_df.sample(25)
```

```python
# Check Record Count (should = 5680 rows)
len(full_df.index)
```

```python
# Text to columns CBSA to seperate State
full_df[['City', 'State']] = full_df['CBSA'].str.split(',',expand=True)
full_df.head(10)
```

```python
# Determine 5 highest "Median AQI" by "State" for each "Year"
```

```python
# 2012
twentytwelve = full_df.loc[full_df['Year'] == 2012].sort_values(by='Median AQI', ascending=False).iloc[:15]
twentytwelve
```

```python
# 2013
twentythirteen = full_df.loc[full_df['Year'] == 2013].sort_values(by='Median AQI', ascending=False).iloc[:15]
```

```python
# 2014
twentyfourteen = full_df.loc[full_df['Year'] == 2014].sort_values(by='Median AQI', ascending=False).iloc[:15]
```

```python
# 2015
twentyfifteen = full_df.loc[full_df['Year'] == 2015].sort_values(by='Median AQI', ascending=False).iloc[:15]
```

```python
# 2016
twentysixteen = full_df.loc[full_df['Year'] == 2016].sort_values(by='Median AQI', ascending=False).iloc[:15]
```

```python
# 2017
twentyseventeen = full_df.loc[full_df['Year'] == 2017].sort_values(by='Median AQI', ascending=False).iloc[:15]
```

```python
# 2018
twentyeighteen = full_df.loc[full_df['Year'] == 2018].sort_values(by='Median AQI', ascending=False).iloc[:15]
```

```python
# 2019
twentynineteen = full_df.loc[full_df['Year'] == 2019].sort_values(by='Median AQI', ascending=False).iloc[:15]
```

```python
# 2020
twentytwenty = full_df.loc[full_df['Year'] == 2020].sort_values(by='Median AQI', ascending=False).iloc[:15]
```

```python
# 2021
twentyone = full_df.loc[full_df['Year'] == 2021].sort_values(by='Median AQI', ascending=False).iloc[:15]
```

```python
# 2022
twentytwo = full_df.loc[full_df['Year'] == 2022].sort_values(by='Median AQI', ascending=False).iloc[:15]
```

```python
# Concate dataframes for Top 15 records for each year (above dfs combined)
top_15_df = pd.concat([twentytwelve, twentythirteen, twentyfourteen, twentyfifteen, twentysixteen, twentyseventeen, twentyeighteen, twentynineteen, twentytwenty, twentyone, twentytwo])
top_15_df
```

```python
# Top average most polluted states from 2012-2022
top_avg_state = top_15_df.groupby("State")["Median AQI"].mean().sort_values(ascending=False)
top_avg_state
```
