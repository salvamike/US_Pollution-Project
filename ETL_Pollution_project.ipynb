```python
## Prepare all 4 AQIs against state and date
US_Pollution_2 = US_Pollution[['State','Date Local','NO2 AQI','O3 AQI','SO2 AQI','CO AQI']]
# Delete rows with NAs
US_Pollution_2 = US_Pollution_2.dropna(axis='rows')  
# Change date from string to date value
US_Pollution_2['Date Local'] = pd.to_datetime(US_Pollution_2['Date Local'],format='%Y-%m-%d')
# Take mean values if there are duplicated entries
US_Pollution_2 = US_Pollution_2.groupby(['State','Date Local']).mean()  
US_Pollution_2_Grouped = US_Pollution_2.groupby(level=0)
```
