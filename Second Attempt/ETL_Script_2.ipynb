```python
# Import Dependencies
import pandas as pd
import psycopg2
from sqlalchemy import create_engine
```

```python
# Create an engine instance
engine = create_engine('postgresql://team_members:us_pollution@localhost:5432/US_Pollution_Project')
```

```python
# Connect PostgreSQL "Master" Database
dbConnection = engine.connect();
```

```python
# Read data from PostgreSQL database table and load into a DataFrame instance
full_df = pd.read_sql("select * from master", dbConnection)
```

```python
# Determine Top 5 Worst AQI Cities
Worst_AQI = full_df.groupby("CBSA")["Median AQI"].mean().sort_values(ascending=False)
Worst_AQI.head(5)
```
